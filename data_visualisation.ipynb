### 2. Histogram

```python
# Histogram of MPG
# Create histograms of MPG in the city and on the highway
plt.figure(figsize=(10, 6))
plt.hist(df['MPG.city'], bins=20, alpha=0.5, color='blue', label='MPG in City')
plt.hist(df['MPG.highway'], bins=20, alpha=0.5, color='green', label='MPG on Highway')

# Add labels and title
plt.xlabel('Miles per Gallon (MPG)')
plt.ylabel('Frequency')
plt.title('Histogram of MPG in the City vs. on the Highway')
plt.legend()

# Show plot
plt.grid(True)
plt.show()
```

### 3. Lineplot

```python
# Lineplot (Wheelbase vs turning circle)

# Creating the line plot
plt.figure(figsize=(10, 6))
plt.plot(df['Wheelbase'], df['Turn.circle'], marker='o', linestyle='-')

# Adding labels and title
plt.xlabel('Wheelbase')
plt.ylabel('Turning Circle')
plt.title('Relationship between Wheelbase and Turning Circle')

# Showing grid
plt.grid(True)

# Showing plot
plt.show()
```

### 4. Barplot

```python
# Barplot for mean horsepower
# Calculate the mean horsepower for each car type
mean_horsepower_by_type = df.groupby('Type')['Horsepower'].mean()

# Create the bar plot
plt.figure(figsize=(10, 6))
mean_horsepower_by_type.plot(kind='bar', color='skyblue')

# Adding labels and title
plt.xlabel('Car Type')
plt.ylabel('Mean Horsepower')
plt.title('Mean Horsepower by Car Type')

# Showing plot
plt.grid(True)
plt.show()
```
